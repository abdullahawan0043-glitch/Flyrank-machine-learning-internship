## 2. Target or proxy

*What would you predict? Where does that label come from — observed outcome or a defined rule?*

```python
"""
Target/proxy: is_declining_label, defined as trend_direction == "down"

This is a beginner proxy label, not the ideal target - it is calculated from the CURRENT window's trend bucket, not a future outcome. A stronger version (planned for later weeks) would instead use a future-looking outcome, such as: features from the prior 90 days -> decline over the next 30 days. Using a proxy now lets me build and test the ranking workflow honestly, while being clear that the real capstone target should be an observed future outcome, not a current-window bucket.
"""
```

## 3. Success metric

*One metric you can defend. What number means 'good'?*

```python
"""
Success metric: Precision@50 (and Precision@K more generally)

Since the output is a ranked list and reviewers only have capacity for a limited number of pages per week, the right metric asks: "of the top K pages the system ranks first, how many are actually true positives (real declining pages)?" This directly matches the real action - reviewing pages in order - unlike a metric like overall accuracy, which would treat all 30,000 pages equally even though only the top of the list ever gets acted on.
"""
```

## 5. Why ML beats a fixed rule here

*What makes the pattern too messy for an if-statement?*

```python
"""
A fixed rule (like "flag any page whose trend is down") already exists as our baseline, and it only reaches Precision@50 = 0.240 (about 12 of the top 50 correct). The learned random forest model reaches Precision@50 = 0.740 (about 37 of the top 50 correct) on the same data - roughly a 3x improvement. This gap shows the pattern behind "which pages are really at risk" is not captured by one or two hand-written thresholds; it depends on many signals (visibility, freshness, position, engagement) interacting in ways that are real but too tangled to write by hand. That is exactly the situation where ML earns its place over a plain rule.
"""
```
