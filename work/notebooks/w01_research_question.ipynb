## 2. The question: decision, action, cost of a wrong call

*What decision does your work improve? Who acts on it? What does a wrong recommendation cost?*

```python
"""
This work improves the decision: which content pages should an editor review and refresh first, given limited time. The person acting on it is a content editor or SEO strategist who has capacity to review only a handful of pages per week - the ranked output tells them where to start.

A wrong recommendation costs in two directions: if the model flags a page that isn't really declining, the editor wastes review time on a page that didn't need it (a false positive). If the model misses a page that IS genuinely declining with real demand, that page keeps losing visibility unnoticed until the loss compounds (a false negative, the costlier mistake since it delays a real fix). Because reviewer time is limited, the ranking must be trustworthy enough that the top of the list is worth acting on first.
"""
```
